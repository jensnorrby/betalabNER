```python
import kblab 
from kblab import Archive
import json
```

```python
class WordSum:
  """
  A class that represents all words in the betalab database.

  Methods
  -------
  survey(titles, years):
    Stores the words from a given number of titles over a given period of time in .txt file. 
  """
  def __init__(self):
    """
    Constructs the necessary parameters and setup for retrieving betalab data.

    Parameters
    ----------
    a : Archive
        Stores the information needed to download betalab data.
    """
    self.a = ""

    with open(r'Betalab\api_credentials.txt', 'r') as file:
      pw = file.read().replace('\n', '')

    #I store the information in self.a to retrieve later.
    self.a = Archive("https://betalab.kb.se", auth=("demo", pw)) 
    
    pass

  def survey(self, titles, years):
    """
   A function to store the words from a list of publications in correspondance with a list of years.
  
   Parameters
   ----------
   titles : list
        A list of titles corresponding to publications in the betalab database.
   years : list
        A list of years covering the period that is studied.

   Returns
   -------
   None
    """
    for year in years:
      for title in titles:
        word_sum = ""
        for package_id in self.a.search(f'label: "{title}" meta.created: "{year}"'):
            package = self.a.get(package_id)

            if "content.json" in package:
                content_json = json.load(package.get_raw("content.json"))
                for item in content_json:
                    for key, val in item.items():
                        if key == "content":
                            word_sum += val

        #The words a stored as a string in a .txt file categorised by title and year, as to simplify the NER step.
        with open(fr'~\Words\{year}\Word_sum {title}.txt', 'w') as convert_file:
            convert_file.write(json.dumps(word_sum))
      print(f"{year} done!")
    pass
```

```python
#These lists correspond to the information I would be interested in retrieving from you now, even if it is not available from betalab.
title_list = ["Aftonbladet", "Dagens Nyheter", "Göteborgsposten", "Arbetet", "Göteborgs Handels- och Sjöfartstidning", "Haparandabladet", "Norbottens-Kuriren", "Stockholms Dagblad", "Svenska Dagbladet", "Sydsvenska Dagbladet"]
year_list = [str(year) for year in range(1939, 1990)]

setup = WordSum()

setup.survey(title_list, year_list)


                        
```
